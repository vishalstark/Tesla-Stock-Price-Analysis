```python
#Data Set link - https://www.kaggle.com/rpaguirre/tesla-stock-price
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact
%matplotlib notebook
```

```python
plt.rcParams["axes.grid"] = False
```

```python
#Reading and Cleaning data

stock=pd.read_csv("Tesla.csv")
stock['Date'] = pd.to_datetime(stock['Date'])

#Yearly date range
tesla_p = { 'p_name' : ['Tesla Model S','Model X','Model 3'], 'l_date' : ['06-22-2012','09-29-2015','04-01-2016']}

fig, ax1 = plt.subplots(figsize=(10,3))
ax1.plot(stock['Date'],stock['High'])
ax1.plot(stock['Date'],stock['Low'])
ax1.plot(stock['Date'],stock['Open'])
ax1.plot(stock['Date'],stock['Close'])
ax1.set_xlabel('Year')
ax1.set_ylabel('In Dollars')
ax1.tick_params('n')
ax1.legend(loc='upper left',fontsize=9,frameon=True,framealpha=1)


ax2 = ax1.twinx()

ax2.plot(stock['Date'],stock['Volume'],alpha=0.5)
ax2.set_ylabel('Volume')
ax2.tick_params('y')
ax2.legend(loc=9,fontsize=9,frameon=True,framealpha=1)
ax2.title.set_text('Tesla Stock Price Trends with Volume')
def f(Volume):
   
    if Volume:
        ax2.set_visible(True)
        ax1.title.set_text('')
    else:
        ax2.set_visible(False)
        ax1.title.set_text('Tesla Stock Price Trends')
interact(f,Volume=True)
fig.tight_layout()
plt.show()
```
